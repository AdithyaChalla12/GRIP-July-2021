### Data Visualization

### Correlation Heat Map

```python
# correlation matrix and heatmap
sns.heatmap(matches.corr(), annot=True, cmap='RdYlGn')
plt.title('Correlation between variables')
```

### Most number of Man of the match awards

```python
plt.figure(figsize=(10,5))
x = matches['player_of_match'].value_counts().values[:20]
y = matches['player_of_match'].value_counts().index[:20]
sns.barplot(y,x)
plt.xticks(rotation=90)
plt.show()
```

### Highest no of Team WIns

```python
plt.figure(figsize=(10,5))
x = matches['winner'].value_counts().values[:20]
y = matches['winner'].value_counts().index[:20]
sns.barplot(y,x)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,10))
matches['winner'].value_counts().plot.pie(autopct="%1.1f%%")
```

### Most no of matches played in stadium

```python
plt.figure(figsize=(10,5))
x = matches['venue'].value_counts().values[:20]
y = matches['venue'].value_counts().index[:20]
sns.barplot(y,x)
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,5))
x = deliveries['batsman_runs'].value_counts().values[:10]
y = deliveries['batsman_runs'].value_counts().index[:10]
sns.barplot(y,x)
plt.xticks(rotation=90)
plt.show()
```

### Highest run Scorer

```python
top_scorers = deliveries.groupby(["match_id", "batsman","batting_team"])["batsman_runs"].sum().reset_index()
top_scorers.sort_values('batsman_runs', ascending=0).head(10)
top_scorers.nlargest(10,'batsman_runs')
```
